```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import os
```

```python
categorias = []
labels = []
imagenes = []
```

```python
#Carpetas donde estan alojadas las imagenes de las letras
categorias = os.listdir("C://Users//RodrigoUnU//Desktop//RN//Images//Images//")
```

```python
#Opcional: visualizar que si estemos donde se encuentran las carpetas
print(categorias)
```

```python
#Pre-procamiento de las imagenes
#Se busca convertir las imagenes en una dimension especifica y posteriormente pasarlas a un arreglo

x=0
for directorio in categorias:
	for imagen in os.listdir("C://Users//RodrigoUnU//Desktop//RN//Images//Images//" + directorio):
		img = Image.open("C://Users//RodrigoUnU//Desktop//RN//Images//Images//" +directorio+"//"+imagen).resize((200,200))
		img = np.asarray(img)
		imagenes.append(img)
		labels.append(x)
	x +=1
```

```python
#Opcional: visualizar las categorias formadas apartir del pre-procesamiento
print(labels)
```

```python
#Convertir el arreglo de imagenes en un numpy
imagenes = np.asanyarray(imagenes)
imagenes.shape
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(200,200)),
    tf.keras.layers.Dense(2000, activation='relu'),
    tf.keras.layers.Dropout(0.00020),
    tf.keras.layers.Dense(500, activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
labels = np.asarray(labels)
```

```python
print(labels)
```

```python
model.fit(imagenes,labels,epochs=200)
```

```python
#Evaluamos el modelo
im = Image.open('C://Users//RodrigoUnU//Desktop//RN//test//test.png').resize((200,200))
im = np.asarray(im)
im = im[:,:,0]
im=np.array([im])
im.shape
test = im
test.shape
```

```python
#Empezamos a realizar predicciones con el modelo ya evaluado
predicciones = 0
predicciones = model.predict(test)
print(predicciones)
```
